```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

# Load your localized dataset
localized_df = pd.read_csv("localized_df.csv")

# Clean coordinates
localized_df['latitude'] = localized_df['latitude'].astype(str).str.replace(',', '').astype(float)
localized_df['longitude'] = localized_df['longitude'].astype(str).str.replace(',', '').astype(float)
localized_df['speed_mps'] = localized_df['speed_mps'].astype(float)

# Pick 5 slowest points as pseudo stops
stops_df = localized_df.nsmallest(5, 'speed_mps').copy()
stops_df.reset_index(drop=True, inplace=True)

# Reverse geocoding
geolocator = Nominatim(user_agent="stop_name_locator")
reverse_geocode = RateLimiter(geolocator.reverse, min_delay_seconds=1)

def get_stop_name(lat, lon):
    try:
        location = reverse_geocode((lat, lon), exactly_one=True)
        if location:
            return location.address.split(',')[0]  # short name
        else:
            return "Unknown Stop"
    except:
        return "Unknown Stop"

stops_df['stop_name'] = stops_df.apply(lambda row: get_stop_name(row['latitude'], row['longitude']), axis=1)

stops_df.to_csv("stops_with_names.csv", index=False)
print(stops_df)
```

```python
import pandas as pd
from geopy.distance import geodesic

# -----------------------------
# Step 1: Load localized dataset
# -----------------------------
localized_df = pd.read_csv("/content/localized_df.csv")  # your GPS + video + localization dataset

# Ensure numeric columns are correctly typed
for col in ['latitude', 'longitude', 'speed_mps']:
    localized_df[col] = localized_df[col].astype(str).str.replace(',', '').astype(float)

# -----------------------------
# Step 2: Calculate distances
# -----------------------------
distances = [0]  # first point distance = 0
for i in range(1, len(localized_df)):
    prev = (localized_df.iloc[i-1]['latitude'], localized_df.iloc[i-1]['longitude'])
    curr = (localized_df.iloc[i]['latitude'], localized_df.iloc[i]['longitude'])
    dist = geodesic(prev, curr).meters
    distances.append(dist)

localized_df['distance_m'] = distances
localized_df['cumulative_distance_m'] = localized_df['distance_m'].cumsum()

# -----------------------------
# Step 3: Compute travel times
# -----------------------------
localized_df['time_sec'] = localized_df['distance_m'] / localized_df['speed_mps']  # seconds
localized_df['cumulative_time_sec'] = localized_df['time_sec'].cumsum()
localized_df['time_diff_sec'] = localized_df['time_sec'].diff().fillna(0)

# -----------------------------
# Step 4: Load detected stops
# -----------------------------
stops_df = pd.read_csv("/content/stops_with_names.csv")  # output from stop detection code
# Example columns: latitude, longitude, stop_name

# -----------------------------
# Step 5: Compute ETA to stops
# -----------------------------
stop_eta_sec = []
for idx, stop in stops_df.iterrows():
    # Find nearest GPS point in localized_df
    distances_to_stop = localized_df.apply(
        lambda row: geodesic(
            (row['latitude'], row['longitude']),
            (stop['latitude'], stop['longitude'])
        ).meters, axis=1
    )
    nearest_idx = distances_to_stop.idxmin()
    eta_sec = localized_df.iloc[nearest_idx]['cumulative_time_sec']
    stop_eta_sec.append(eta_sec)

stops_df['eta_sec'] = stop_eta_sec
stops_df['eta_min'] = stops_df['eta_sec'] / 60  # convert seconds to minutes
stops_df['stop_number'] = ["Stop " + str(i+1) for i in range(len(stops_df))]

# -----------------------------
# Step 6: Merge stops with route
# -----------------------------
route_planning_df = localized_df.merge(
    stops_df[['latitude', 'longitude', 'stop_name', 'eta_min', 'stop_number']],
    on=['latitude', 'longitude'],
    how='left'
)

# -----------------------------
# Step 7: Save final route planning
# -----------------------------
route_planning_df.to_csv("route_planning.csv", index=False)
print("Route planning completed successfully. Sample output:")
print(route_planning_df.head(20))
```

```python
import pandas as pd
from gtts import gTTS
from IPython.display import Audio, display
import folium
from geopy.distance import geodesic

# -----------------------------
# Step 1: Load Route Planning CSV
# -----------------------------
route_df = pd.read_csv("/content/route_planning.csv")

# Ensure numeric types
for col in ['latitude', 'longitude', 'speed_mps', 'eta_min']:
    if col in route_df.columns:
        route_df[col] = pd.to_numeric(route_df[col], errors='coerce')

# -----------------------------
# Step 2: Audio Announcements
# -----------------------------
# Announcement for next stop
def announce_stop(row):
    if pd.notnull(row['stop_name']):
        message = f"Attention passengers. Approaching {row['stop_name']}. ETA {int(row['eta_min'])} minutes."
        tts = gTTS(text=message, lang='en')
        tts.save("stop_announcement.mp3")
        display(Audio("stop_announcement.mp3", autoplay=True))

# Example: Announce for all stops
stops_rows = route_df[route_df['stop_name'].notnull()]
for idx, stop in stops_rows.iterrows():
    announce_stop(stop)

# -----------------------------
# Step 3: Automatic Door Opening Simulation
# -----------------------------
# We'll trigger "door open" if speed is 0 or very low near a stop
def check_door(row, threshold_speed=1.0):
    if pd.notnull(row['stop_name']) and row['speed_mps'] <= threshold_speed:
        print(f"Door opening at {row['stop_name']} (Lat: {row['latitude']}, Lon: {row['longitude']})")

for idx, row in route_df.iterrows():
    check_door(row)

# -----------------------------
# Step 4: Route Visualization using Folium
# -----------------------------
# Center map at the first point
start_lat = route_df.iloc[0]['latitude']
start_lon = route_df.iloc[0]['longitude']
route_map = folium.Map(location=[start_lat, start_lon], zoom_start=15)

# Add GPS points (trajectory)
for i, row in route_df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=3,
        color='blue',
        fill=True
    ).add_to(route_map)

# Add stops
stops_only = route_df[route_df['stop_name'].notnull()]
for i, stop in stops_only.iterrows():
    folium.Marker(
        location=[stop['latitude'], stop['longitude']],
        popup=f"{stop['stop_name']} (ETA: {int(stop['eta_min'])} min)",
        icon=folium.Icon(color='red', icon='bus', prefix='fa')
    ).add_to(route_map)

# Save map
route_map.save("route_map.html")
print("Route visualization saved as route_map.html. Open it in browser to view.")
```
